```python
import os
import pandas as pd
```

```python
data = pd.read_csv("100 Sales Records.csv")
data.head()
```

```python
nan_df= data[data.isna().any(axis=1)]
display(nan_df.head(10))
```

```python
data['Year'] = data['Order Date'].str[6:12]
data['Year'] = data ['Year'].astype('int32')
data.head(30)
```

```python
import matplotlib.pyplot as plt
Year = range(10,18)
print(Year)
plt.bar(x= data['Year'], height= data['Total Profit'])
plt.xticks(Year)
plt.ylabel('Profit in USD $')
plt.xlabel('Year Number')
plt.show()
```

```python
data.groupby(['Country']).sum()
```

```python
import matplotlib.pyplot as plt
keys= [Country for Country, df in data.groupby(['Country'])]
plt.bar(keys, data.groupby(['Country']).sum()['Total Profit'])
plt.ylabel('Profit in USD $', size=20)
plt.xlabel('Year Number', size=20)
plt.xticks(keys, rotation= 'vertical', size=12)
plt.rcParams['figure.figsize'] = [10, 5]
plt.show()
```

```python
import matplotlib.pyplot as plt
keys= [Item_Type for Item_Type, df in data.groupby(['Item Type'])]
plt.bar(keys, data.groupby(['Item Type']).sum()['Units Sold'])
plt.ylabel('Units Sold', size=20)
plt.xlabel('Item Type', size=20)
plt.xticks(keys, rotation= 'vertical', size=12)
plt.rcParams['figure.figsize'] = [20, 5]
plt.show()
```

```python
prices = data.groupby('Item Type').mean()['Unit Price']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, data.groupby(['Item Type']).sum()['Units Sold'], color='b')
ax2.plot(keys, data.groupby(['Item Type']).sum()['Units Sold'], color='g')

ax1.set_xlabel('Item Type', size= 20)
ax1.set_ylabel('Units Sold', size= 20)
ax2.set_ylabel('Price ($)', size= 20)
ax1.set_xticklabels(keys, rotation= 'vertical', size= 15)

fig.show()
```
